```python
import pandas as pd
import requests, json

base_url = "https://fantasy.premierleague.com/api/"

pd.set_option('display.max_columns', None)
```

```python
fpl_static_data_json = requests.get(base_url + "bootstrap-static/").json()
fpl_static_data = pd.json_normalize(fpl_static_data_json)
```

```python
# Current season == 2024/25
# 'elements' field contains data for all premier league players in the current FPL season
current_season_player_data_df = pd.DataFrame(fpl_static_data['elements'])
teams_df = pd.DataFrame(fpl_static_data['teams'])
positions_df = pd.DataFrame(fpl_static_data['element_types'])
# Replace the element_type (player positions) and team ids with their respective names for better readability
current_season_player_data_df['element_type'] = current_season_player_data_df['element_type'].replace(positions_df.set_index('id')['singular_name_short'])
current_season_player_data_df['team'] = current_season_player_data_df['team'].replace(teams_df.set_index('id')['name'])
```
